# 1. Bank Account Manager

```python
# Account Information
account_holder="Reena Gupta"
account_number="5800123678935"
balance=5000.0 # intial balance

#Functions and Conditions

#Function for deposit
def deposit(amount1):
  global balance
  if amount1>0:
    balance= balance + amount1
    print(f"Deposit of Rs{amount1} to account Successful! New Balance is{balance}")
  else:
    print("Invalid deposit amount.Must be greater than 0")

#Function for withdrawal
def withdraw(amount2):
  global balance
  if amount2<=0:
    print("Invalid withdrawal amount.")
  elif amount2>balance:
    print("Insufficient balance for withdrawal")
  else:
    balance=balance-amount2
    print(f"Withdrawal of Rs{amount2} Successful! New Balance is{balance}")

#Function to check balance
def check_balance():
  print(f"Current balance:Rs{balance}")

# Enquiry Choices
print("Welcome to the Bank Account Manager")
print(f"Account Holder:{account_holder}")
print(f"Account Number:{account_number}")

while True:
  print("\n Choose an option")
  print("1. Deposit")
  print("2. Withdrawal")
  print("3. Check Balance")
  print("4. Exit")
  choice=input("Enter your choice(1-4):")
  if choice=='1':
    amt= float(input("Enter the amount to be deposited: Rs"))
    deposit(amt)
  elif choice=='2':
    amt= float(input("Enter the amount to withdraw: Rs"))
    withdraw(amt)
  elif choice=='3':
    check_balance()
  elif choice=='4':
    print("Thank you for using the Bank Account Manager")
    break
  else:
    print("Invalid choice.Please select a option between (1-4)")
```

# 2. Weather Data Analyzer

```python
# Taking data from User
region=input("Enter the name of region:")
num_days=int(input(f"How many days of weather data for {region}?"))

# Initializing empty lists to add data
temperatures=[]
humidity=[]

# Getting temperature and humidity for each day
for day in range(1,num_days+1):
  temp= float(input(f"Enter temperature for Day{day}(°C):"))
  hum= float(input(f"Enter humidity for Day{day}(%):"))
  temperatures.append(temp)
  humidity.append(hum)

# Function to determine weather condition
def determine_weather_condition(temp,hum):
  if temp>30 and hum>70:
    return "Hot and Humid"
  elif temp>=30:
    return "Sunny"
  elif temp<20:
    return "Cold"
  elif hum>75:
    return "Rainy"
  else:
    return "Mild"

# Function to calculate average of data
def calculate_average(data):
  return sum(data)/len(data)

# Calculating average
avg_temp= calculate_average(temperatures)
avg_hum= calculate_average(humidity)

# Display results
print(f"\n Weather Report for {region}:\n")
for i in range(num_days):
  condition= determine_weather_condition(temperatures[i],humidity[i])
  print(f"Day{i+1}: Temperature= {temperatures[i]}°C , Humidity= {humidity[i]}% --> Condition= {condition}")

# Printing Averages
print(f"\n Average Temperature: {round(avg_temp,2)}°C")
print(f" Average Humidity: {round(avg_hum,2)}%")
```
