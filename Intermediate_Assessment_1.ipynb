The above codes shows the difference between global and local variable
- global variable is the variable that can be accessed anywhere within the program including inside the functions
- local variable are those which can be accessed only within the function  

Here dis_rate1 (discount rate) is declared outside the function and is global and hence the print statement inside and outside the function got printed and the operations inside the function could be performed.Similarly dis_rate3 is also a global variable declared inside the function and could be accessed even outside to print it.

But dis_rate2 is a local variable which was declared only inside the function and hence only the print statement inside function got printed and when we tried accessing it outside function to print, an error occurred as its not defined and hence using that variable no operations can be performed and is only exclusive for the condition inside function.

```python
# custom iterator class to yield one order at a time
class OrderIterator:
  def __init__(self,orders):
    self.orders= orders # store the list of orders
    self.index=0 # start index
  def __iter__(self): # returns a iterator object
    return self
  def __next__(self):
    if self.index < len(self.orders):
      order= self.orders[self.index]
      self.index+=1
      return order
    else:
      raise StopIteration

order_iter=OrderIterator(sales_data)
print(order_iter.__next__()) # priniting the 1st order
```

```python
for order in order_iter: # priniting all other orders which was called by the class one by one using loop
    print(order)
```

```python
# grouping using dictionary comprehension
unique_date = {sale['order_date'] for sale in sales_data}

grouped_by_orderdate = {
    date: [sales for sales in sales_data if sales['order_date'] == date]
    for date in unique_date
}

for date,orders in sorted(grouped_by_orderdate.items()):
  print(f"\n Date:{date}")
  for order in orders:
    print(f"Order ID: {order['order_id']}, Customer: {order['customer']},Item: {order['item']}, Quantity: {order['quantity']}, Price: {order['price_per_item(Rs)']}")
```
