## 2. Filtering

## 3. Statistical Analysis

```python
weightArray=np.array(data['weight'])
print("25th percentile of weight:", np.percentile(weightArray, 25))
print("50th percentile (Median) of weight:",np.percentile(weightArray, 50) )
print("75th percentile of weight:", np.percentile(weightArray, 75))
```

## 4. Array Manipulation

```python
accelArray=np.array(data['acceleration'])
normalized= (accelArray - accelArray.min())/(accelArray.max()-accelArray.min())
print("Normalized acceleration values:", normalized)
```

## 5. Broadcasting

```python
hpArray=np.array(data['horsepower'])
# in the array wherever there is a ? it has to be stored as np.nan indicating its not a number for which we can check and replace by mean
# hence the number is checked and replace() function ignores the appearance of . or '' in number once and then will check if its a digit and returns if its true or false as if . appears twice its anyway not a valid number
# if the condition is true the number will be maintained as its and converted to float else will be stored as np.nan
hp_cleaned = np.array([
    float(x) if str(x).replace('.', '', 1).isdigit() else np.nan
    for x in hpArray
])
mean_hp=np.nanmean(hp_cleaned) # it computes the mean ignoring the non number values
hp_filled = np.where(np.isnan(hp_cleaned), mean_hp, hp_cleaned) # isnan() checks if it nan or a number
#np.where tells the index where it has be replaced and replaces,
#if its true then it replaces with mean else it maintains the same number in hp_cleaned
hp_new = hp_filled + (hp_filled*(0.1))
print("Horse power increased by 10%:",hp_new)
```

## 6. Boolean Indexing

```python
disArray=np.array(data['displacement'])
originArray=np.array(data['origin'])
europe_origin = originArray == 2 #boolean mask condition
europe_displacement=disArray[europe_origin] # gives the corresponding displacement where origin is true ie =2
print("Average displacement of European cars:",  np.mean(europe_displacement))
```

## 8. Sorting

```python
model_yearArray= np.array(data['model year'])
sorted_indices= np.argsort(-model_yearArray) # sorts the indices and give the no. of index array and the model year value being sorted ie the index of last row comes first 397,396 like that
sorted_cars=car_namesArray[sorted_indices] # it will give the car names corresponding to the sorted index in descending order for model year
print("Top 5 newest cars by model year(82):")
for name in sorted_cars[:5]:
    print(name)
```

## 9. Correlation

Pearson correlation coefficient= cov(x,y)/std(x)std(y)


```python
corr_matrix = np.corrcoef(mpgArray, weightArray)
print("Pearson correlation coefficient between mpg and weight is ",corr_matrix[0,1]) # [0,1] since coocoef gives array the off diagonal element is the value
```

## 10. Conditional Aggregates

```python
unique_cyl= np.unique(cylindersArray)
print("Unique cylinders:",unique_cyl)
for cyl in unique_cyl:
  unique_mpg= mpgArray[cylindersArray== cyl]
  mean_mpg= np.mean(unique_mpg)
  print(f"Mean of mpg for cylinder {cyl} is",mean_mpg)
```

## 2. Filtering and Indexing

```python
logical_expression= (data['weight']<3000) & (data['model year']==75)
print("Cars with weight less than 3000 in the year 1975:")
data[logical_expression][['car name','weight','mpg']]
```

## 3. Handling Missing Data

```python
data=data.replace('?', pd.NA)
data['horsepower'] = pd.to_numeric(data['horsepower'],errors='coerce')
print("Before filling:\n", data.isnull().sum())
data['horsepower'] = data['horsepower'].fillna(data['horsepower'].median())
print("\nAfter filling:\n", data.isnull().sum())
```

## 4. Data Transformation

```python
data['power_to_weight_ratio']=data['horsepower']/data['weight']
data.head()
```

## 5. Group By

```python
g=data.groupby('origin')['mpg'].mean()
g
```

## 6. Sorting

```python
data.sort_values(by='mpg',ascending=False)[['car name','mpg']][0:10]
```

## 7. Apply Function

```python
def performance_score(row):
  return row['mpg'] * row['acceleration'] / row['weight']   # row no['column name']=value there
data['performance_score'] = data.apply(performance_score, axis=1) # after applying for each row it should be created as a column named performance_score
data.head()
```

## 8. Visualization Preparation

```python
data.groupby('model year')[['mpg','weight','horsepower']].mean()
```

## 10. Finding Anomalies



*   




```python
Q1 = data['mpg'].quantile(0.25)
Q3 = data['mpg'].quantile(0.75)
IQR = Q3 - Q1
print("IQR for mpg is",IQR)
```



*   


```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_outliers = data[(data['mpg'] < lower_bound)]
upper_outliers=data[(data['mpg'] > upper_bound)]
outliers= pd.concat([lower_outliers,upper_outliers ],\         #new data frame outliers which stores the lower and upper outliers
                             ignore_index=True)
```



*   



```python
outlier_cars = outliers[['car name', 'mpg', 'model year']] #new data frame
outlier_cars
```
